```python
# pizza need
pizza=6
each=4
num=int(input())
need=num*each
for i in range(num):
    if need%pizza==0:
        print("You need to buy: ",need//pizza)
        print("left pizza slice", (pizza*(need//pizza))-need)
        break
    else:
        print("You need to buy: ",(need//pizza)+1)
        print("left pizza slice", (pizza*((need//pizza)+1))-need)
        break
```
