```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y,c="r")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("x versus y")
```

```python
# plt.subplot(nrows, ncols, plot_number
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,y)
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Title")
#insert secont figure
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(x,y)
```

```python
fig=plt.figure(figsize=(10,5), dpi=80)
#figsize is a tuple of the width and height of the figure in inches
#dpi is the dots-per-inch (pixel per inch)
axes=fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y,'ro',label="x vs y")
axes.plot(x,y**2, "go", label=" x vs y*3")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.legend(loc=0)
```
