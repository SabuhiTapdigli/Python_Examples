```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("BankCustomers.csv")
Main=df.iloc[:,3:]
Geography=pd.get_dummies(Main["Geography"],drop_first=True)
Gender=pd.get_dummies(Main["Gender"],drop_first=True)
Main=pd.concat([Main,Geography],axis=1)
Main=pd.concat([Main,Gender],axis=1)
Main.drop(["Geography","Gender"],axis=1,inplace=True)
x=Main.iloc[:,1:]
y=Main.iloc[:,0:1]
```

```python
#Normalization
x=x.apply(lambda x: (x-min(x))/(max(x)-min(x)))
y=y.apply(lambda y: (y-min(y))/(max(y)-min(y)))
```
